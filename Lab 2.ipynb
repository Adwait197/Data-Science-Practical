```python
abc=df['Sem1'].values.flatten()
```

```python
abc
```

```python
mean = np.mean(abc)
mean
```

```python
Stdeviation = np.std(abc)
Stdeviation
```

```python
threshold = 3
outlier = []
for i in abc:
    z = (i-mean)/Stdeviation
    if z > threshold:
        outlier.append(i)
print("Outliers in Datasets are : ",outlier)
```

```python
for i in outlier:
    arr = np.where(abc == i)
    print(arr)
    
```

```python
# IQR

DF_IQR=df1.copy()
```

```python
sorted(DF_IQR['Sem5'])
```

```python
#flooring and capping
Q1=np.quantile(DF_IQR['Sem5'],0.25)
Q3=np.quantile(DF_IQR['Sem5'],0.75)
IQR = Q3 - Q1
print("Q1 : ",Q1)
print("Q3 : ",Q3)
print("IQR : ",IQR)
```

```python
lower_bound = Q1-(1.5*IQR)
upper_bound = Q3+(1.5*IQR)
print("Lower Bound : ",lower_bound)
print("Upper Bound : ",upper_bound)
```
